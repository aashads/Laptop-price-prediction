```python
def fetch_processor(text):
  if text == 'Intel Core i5' or text == 'Intel Core i3' or text == 'Intel Core i7':
    return text
  else:
    if text.split()[0] == 'Intel':
        return 'Other Intel Processor'
    else:
        return 'AMD Processor'
```

```python
df['Cpu Brand']=df['Cpu Name'].apply(fetch_processor)
df
```

```python
def cat_os(inp):
    if inp == 'Windows 10' or inp == 'Windows 7' or inp == 'Windows 10 S':
        return 'Windows'
    elif inp == 'macOS' or inp == 'Mac OS X':
        return 'Mac'
    else:
        return 'Others/No OS/Linux'
```

```python
#Function to remove outliers
def remove_outlier(df,col):
  Q1=df[col].quantile(0.25)
  Q3=df[col].quantile(0.75)
  IQR=Q3-Q1
  lower=Q1-1.5*IQR
  upper=Q3+1.5*IQR
  df_out=df.loc[(df[col]>lower)&(df[col]<upper)]
  return df_out
```

```python
#Splitting x_value
x=dfe.drop(['Price'],axis=1)
x
```

```python
#splitting y values
y=df['Price']
y
```

```python
#Splitting Training and Testing Data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
```

```python
#Model creation using Linear Regression
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
y_pred
```

```python
#Linear Regression Model Performance Evaluation
from sklearn.metrics import mean_absolute_percentage_error,r2_score
print(r2_score(y_test,y_pred))
print(mean_absolute_percentage_error(y_test,y_pred))
```

```python
#Hyper Tuning to improve r2_score
from sklearn.model_selection import GridSearchCV
model4=LinearRegression()
param4={'fit_intercept': [True, False], 'n_jobs': [None, 1, -1]}
clf2=GridSearchCV(model4,param4,cv=10,scoring='neg_mean_squared_error')
clf2.fit(x_train,y_train)
print(clf2.best_params_)
```

```python
model5=LinearRegression(fit_intercept=False,n_jobs=None)
model5.fit(x_train,y_train)
y_pred5=model5.predict(x_test)
y_pred5
```

```python
#Accuracy Score After Hyper tuning
print("The r2 score after tuning of linear regression model is",r2_score(y_test,y_pred5))
print("The mean absolute error percentage is",mean_absolute_percentage_error(y_test,y_pred5))
```

```python
#Model creation using KNeighbors algoithm
from sklearn.neighbors import KNeighborsRegressor
model1 = KNeighborsRegressor()
param = {'n_neighbors': [3, 5, 7, 9],'weights': ['uniform','distance']}
clf = GridSearchCV(model1, param, cv=10, scoring='neg_mean_squared_error')
clf.fit(x_train, y_train)
print(clf.best_params_)
```

```python
model2=KNeighborsRegressor(n_neighbors=5,weights='distance')
model2.fit(x_train,y_train)
y_pred1=model2.predict(x_test)
y_pred1
```

```python
print("The r2_score after performing tuning on kNeighbors model",r2_score(y_test,y_pred1))
print("The mean absolute error percentage is",mean_absolute_percentage_error(y_test,y_pred1))
```

```python
#Model creation using SVM
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
model2 = SVR()
param1 = {'C': [1.0, 10], 'kernel': ['linear', 'poly', 'rbf'], 'degree': [3, 5], 'gamma': ['scale', 'auto']}
clf1 = GridSearchCV(model2, param1, cv=10, scoring='neg_mean_squared_error')
clf1.fit(x_train, y_train)
print(clf1.best_params_)
```

```python
model3=SVR(C=10,degree=3,gamma='auto',kernel='poly')
model3.fit(x_train,y_train)
y_pred2=model3.predict(x_test)
y_pred2
```

```python
#Model_Evaluation using SVM Model(r2_score is low compared to linear regression and Kneighbors algorithm)
print(r2_score(y_test,y_pred2))
print(mean_absolute_percentage_error(y_test,y_pred2))
```
